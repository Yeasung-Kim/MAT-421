# Continuity, Differentiation, and Taylor's Theorem with Code Examples


## 1. Continuity

Continuity is a key concept in calculus that ensures that a function does not have abrupt changes, jumps, or holes at a given point.
Mathematically, a function `f(x)` is continuous at a point `x = a` if the limit as `x` approaches `a` is equal to the value of the function at that point.

### Definition:
A function `f` is continuous at a point `x = a` if:
lim_{x â†’ a} f(x) = f(a)

### Example:
The function f(x) = x^2 is continuous for all real numbers because there are no abrupt changes in its behavior.


```python
import numpy as np
import matplotlib.pyplot as plt

# Example function: f(x) = x^2
def f_continuous(x):
    return x**2

x = np.linspace(-2, 2, 100)
y = f_continuous(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.title('Continuous Function Example: f(x) = x^2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```


## 2. Differentiation

Differentiation measures the rate of change of a function with respect to one of its variables.
It is represented as the slope of the tangent line to the curve of the function at a given point.

### Definition:
If f(x) is a function, then its derivative, f'(x), is defined as:
f'(x) = lim_{h â†’ 0} [f(x + h) - f(x)] / h

### Example:
The derivative of f(x) = x^2 is f'(x) = 2x, which represents the slope of the tangent line at each point on the curve.


```python
def f_diff(x):
    return x**2

def f_prime_diff(x):
    return 2 * x

x = np.linspace(-2, 2, 100)
y = f_diff(x)
y_prime = f_prime_diff(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2')
plt.plot(x, y_prime, label="f'(x) = 2x")
plt.title("Function and Its Derivative: f(x) = x^2 and f'(x) = 2x")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## 3. Taylor's Theorem

Taylor's Theorem gives an approximation of a differentiable function around a given point using a polynomial.
It states that if a function f(x) is m-times differentiable, then it can be approximated using a Taylor polynomial.

### Formula:
f(x) â‰ˆ f(a) + (x-a)f'(a) + (1/2!)(x-a)^2 f''(a) + ... + (1/m!)(x-a)^m f^(m)(a)

### Example:
Using Taylor's Theorem to approximate f(x) = e^x around x = 0.

```python
from sympy import symbols, exp, diff

x = symbols('x')
f = exp(x)

taylor_expansion = f + (diff(f, x).subs(x, 0)) * x + (diff(f, x, 2).subs(x, 0)) * x**2 / 2

print(f'Taylor Series Approximation of e^x at x = 0: {taylor_expansion}')
```

## 4. Optimization with Gradient Descent

Gradient Descent is an iterative optimization algorithm used to find the local minimum of a differentiable function.
It works by iteratively moving in the direction of the negative gradient of the function.

### Algorithm:
x_{k+1} = x_k - Î± * âˆ‡f(x_k)

where:
- x_k is the current point,
- Î± is the learning rate,
- âˆ‡f(x_k) is the gradient of the function at x_k.

### Example:
Using gradient descent to find the minimum of f(x) = x^2.

```python
def f_opt(x):
    return x**2

def f_prime_opt(x):
    return 2 * x

x = 10  # Starting point
learning_rate = 0.1
tolerance = 0.001
max_iterations = 100

iterations = 0
while abs(f_prime_opt(x)) > tolerance and iterations < max_iterations:
    x = x - learning_rate * f_prime_opt(x)
    iterations += 1

print(f"Minimum found at x = {x} after {iterations} iterations.")
```
